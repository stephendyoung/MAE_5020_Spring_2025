```python
import pandas as pd
import numpy as np



# https://www.kaggle.com/datasets/vinayak123tyagi/bearing-dataset
# https://github.com/jeffprosise/Machine-Learning/blob/master/Anomaly%20Detection%20(PCA).ipynb
file = "https://raw.githubusercontent.com/jeffprosise/Machine-Learning/refs/heads/master/Data/bearings.csv"

# Read the CSV into a pandas DataFrame, using the first column as a DateTime index
df = pd.read_csv(file, index_col=0, parse_dates=[0])

# Convert the DataFrame to a NumPy array
da = df.to_numpy()

data_array = da[:700,:]

print("Data shape:", data_array.shape)
print("First 5 rows as a NumPy array:\n", data_array)
```

```python
# prompt: plot data_array

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(data_array)
plt.xlabel("Time")
plt.ylabel("Values")
plt.title("Plot of data_array")
plt.grid(True)
plt.show()
```

```python

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

plt.rcParams['figure.figsize'] = [6, 12]
plt.rcParams.update({'font.size': 16})

# SVD
U, S, VT = np.linalg.svd(data_array,full_matrices=0)


#Plot the singular values and cumulative sum of singular values.
fig1 = plt.figure()
ax1 = fig1.add_subplot(211)
ax1.semilogy(S,'-o',color='k')
ax2 = fig1.add_subplot(212)
ax2.plot(np.cumsum(S)/np.sum(S),'-o',color='k')

plt.show()
```

```python
# The first principal component

plt.rcParams['figure.figsize'] = [6, 1]
plt.rcParams.update({'font.size': 16})

fig2 = plt.figure()
ax = fig2.add_subplot(111)

for j in range(data_array.shape[0]):
    x = VT[0,:] @ data_array[j,:].T
    if j <500:
        plt.scatter(x, 0, alpha=0.5, marker='o',color='g',s=20)
    else:
        plt.scatter(x, 0, alpha=0.5, marker='x',color='r',s=20)
plt.show()
```

```python
# The second principal component

plt.rcParams['figure.figsize'] = [6, 1]
plt.rcParams.update({'font.size': 16})

fig2 = plt.figure()
ax = fig2.add_subplot(111)

for j in range(data_array.shape[0]):
    x = VT[1,:] @ data_array[j,:].T
    if j <500:
        plt.scatter(x, 0, alpha=0.5, marker='o',color='g',s=20)
    else:
        plt.scatter(x, 0, alpha=0.5, marker='x',color='r',s=20)
plt.show()
```

```python
# The third principal component

plt.rcParams['figure.figsize'] = [6, 1]
plt.rcParams.update({'font.size': 16})

fig2 = plt.figure()
ax = fig2.add_subplot(111)

for j in range(data_array.shape[0]):
    x = VT[2,:] @ data_array[j,:].T
    if j <500:
        plt.scatter(x, 0, alpha=0.5, marker='o',color='g',s=20)
    else:
        plt.scatter(x, 0, alpha=0.5, marker='x',color='r',s=20)
plt.show()
```

```python
# The first and second principal components

plt.rcParams['figure.figsize'] = [6, 6]
plt.rcParams.update({'font.size': 16})

fig3 = plt.figure()
ax = fig3.add_subplot(111)

for j in range(data_array.shape[0]):
    x = VT[0,:] @ data_array[j,:].T
    y = VT[1,:] @ data_array[j,:].T
    if j < 500:
        ax.scatter(x, y, alpha=0.5, marker='o',color='g',s=20)
    else:
        ax.scatter(x, y, alpha=0.5, marker='x',color='r',s=20)

plt.grid()
plt.xlim((-0.15, -0.10))
plt.ylim((-0.02, 0.02))

plt.show()
```
