```python
import pandas as pd
import numpy as np



# https://www.kaggle.com/datasets/vinayak123tyagi/bearing-dataset
# https://github.com/jeffprosise/Machine-Learning/blob/master/Anomaly%20Detection%20(PCA).ipynb
file = "https://raw.githubusercontent.com/jeffprosise/Machine-Learning/refs/heads/master/Data/bearings.csv"

# Read the CSV into a pandas DataFrame, using the first column as a DateTime index
df = pd.read_csv(file, index_col=0, parse_dates=[0])

# Convert the DataFrame to a NumPy array
da = df.to_numpy()

data_array = da[:700,:]

print("Data shape:", data_array.shape)
print("First 5 rows as a NumPy array:\n", data_array)
```

```python
# Plot data_array
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(data_array)
plt.xlabel("Time")
plt.ylabel("Values")
plt.title("Plot of data_array")
plt.grid(True)
plt.show()
```
